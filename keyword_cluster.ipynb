```python
import pandas as pd
import numpy as np
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn import cluster
import matplotlib.pyplot as plt
from nltk.stem import WordNetLemmatizer
import nltk
import jieba
```

# AffinityPropagation

```python
stemmer = PorterStemmer()
sw = stopwords.words('english')
```

```python
def tokenizer(keyword):
    return [stemmer.stem(w) for w in keyword.split(' ')]
```

```python
queries = ['School', 'school', 'chocolate', 'chcoolate', 'choco late', 'choco', 'schol', 'school bus', 'working', '維他命 b', '維他奶', '維他檸檬茶', '陽光檸檬茶', '檸檬', '牛奶' , 'chocolate 牛奶', '陽光 school']
```

```python
tfidf = TfidfVectorizer(use_idf=False, norm=None)
a = tfidf.fit_transform(queries).toarray()
pd.DataFrame(a, index=queries, columns=tfidf.get_feature_names_out())
```

```python
tfidf = TfidfVectorizer(tokenizer=tokenizer)
a = tfidf.fit_transform(queries).toarray()
X = pd.DataFrame(a, index=queries, columns=tfidf.get_feature_names_out())
c = cluster.AffinityPropagation()
```

```python
c.fit_predict(X)
```

```python
X['pred'] = c.fit_predict(X)
```
